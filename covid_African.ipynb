```python
#import the necessary libraries
import pandas as pd
```

```python
# load the data
covid = pd.read_csv('covid_africa.csv')
```

```python
covid.head()
```

```python
covid.shape
```

# which country has the most cases of covid?

```python
covid.plot(kind='bar',
            x='Country',y='Total Cases',
            figsize=(14, 6),
            grid=True,
            legend=None,
            title='Total cases in Africa countries', 
            ylabel='Total cases')
```

# which country has the highest percentage per popoulation?
                    

### Add a column called "percent/pop" to the data frame

```python
covid['percent/pop']=(covid["Total Cases"]/covid['Population'])*100
```

```python
covid.head()
```

```python
covid.plot(kind='bar',
            x='Country',y='percent/pop',
            figsize=(14, 6),
            grid=True,
            legend=None,
            title='Percentage per population in Africa countries', 
            ylabel='% per population',
            color="r")
             
```
